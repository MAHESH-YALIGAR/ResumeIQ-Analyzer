```python
import pdfplumber

def extract_text(file):
    """Extract all text from a PDF file"""
    text = ""
    with pdfplumber.open(file) as pdf:
        for page in pdf.pages:
            extracted = page.extract_text()
            if extracted:
                text += extracted + "\n"
    return text.strip()
  
  
text=extract_text("chandan_resume_2023.docx.pdf")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def preprocess(text):
    tokens = word_tokenize(text.lower())
    tokens = [t for t in tokens if t.isalpha()]
    tokens = [t for t in tokens if t not in stopwords.words('english')]
    return tokens
text=preprocess(text)
text
```

```python
def extract_skills(text, skilled_db):
    found = []
    all_skills = []
    for category, skills in skilled_db.items():
        all_skills.extend(skills)
    for skill in all_skills:
        if skill in text:
            found.append(skill)
    return found
found = extract_skills(text, skilled_db)

print(found)
```

```python
import pandas  as pd
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
data=pd.read_csv("resume.csv")
resume_text=data['text']
resume_role=data['role']

vectorize=TfidfVectorizer()
x=vectorize.fit_transform(resume_text)
x.toarray()
model=LogisticRegression()
model.fit(x,resume_role)
# model.predict()
```

```python



import numpy as np

description = ["looking for python ai developer"]
lang=['pandas','pytorch']
# convert skills to text
resume_text = " ".join(lang)

resume_vector = vectorize.transform(found)
jd_vec = vectorize.transform(description)

# prediction
prediction = model.predict(resume_vector)
print("Prediction:", prediction)

# probabilities
probabilities = model.predict_proba(resume_vector)


predicted_role_index = np.argmax(probabilities[0])

predicted_role = model.classes_[predicted_role_index]

confidence_score = probabilities[0][predicted_role_index] * 100
probs = probabilities[0]
matching_data=[]
top_indices = np.argsort(probs)[::-1]   # highest → lowest
for i in top_indices[:5]:
    role = model.classes_[i]
    score = probs[i] * 100
    # matching_data.append()
    matching_data.append((role, score)) 

    print(f"{role} → {score:.2f}%")
print(f"Role: {predicted_role}")
print(f"Confidence: {confidence_score:.2f}%")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(resume_vector, jd_vec)
print("Match Score:", similarity[0][0] * 100)
```
